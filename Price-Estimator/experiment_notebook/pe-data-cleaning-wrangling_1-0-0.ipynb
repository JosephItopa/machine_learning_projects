```python
source_li=['KEBBI', 'KANO', 'LAGOS', 'OYO', 'KOGI', 'DELTA', 'RIVERS', 'KADUNA', 'OGUN', 'ANAMBRA',
    'CROSS RIVER', 'PLATEAU', 'KWARA', 'KATSINA', 'ADAMAWA', 'ABIA', 'NIGER', 'FEDERAL CAPITAL TERRITORY',
    'BENUE', 'EDO', 'OSUN', 'BORNO', 'IMO', 'ENUGU', 'EKITI', 'ONDO', 'NASARAWA', 'YOBE', 'ZAMFARA', 'SOKOTO', 
    'GOMBE', 'BAUCHI', 'JIGAWA', 'AKWA IBOM', 'BAYELSA', 'TARABA']

dest_li=['KEBBI', 'KANO', 'LAGOS', 'OYO', 'KOGI', 'DELTA', 'RIVERS', 'KADUNA', 'OGUN', 'ANAMBRA',
    'CROSS RIVER', 'PLATEAU', 'KWARA', 'KATSINA', 'ADAMAWA', 'ABIA', 'NIGER', 'FEDERAL CAPITAL TERRITORY',
    'BENUE', 'EDO', 'OSUN', 'BORNO', 'IMO', 'ENUGU', 'EKITI', 'ONDO', 'NASARAWA', 'YOBE', 'ZAMFARA', 'SOKOTO', 
    'GOMBE', 'BAUCHI', 'JIGAWA', 'AKWA IBOM', 'BAYELSA', 'TARABA']

new_price = new_price[new_price.source.isin(source_li)]
new_price = new_price[new_price.destination.isin(dest_li)]
```

```python
#Feature Engineering of Time Series Column
new_price['loadedDate'] = pd.to_datetime(new_price['loadedDate'], format='%Y-%m-%d')
new_price['year'] = new_price['loadedDate'].dt.year
new_price['month'] = new_price['loadedDate'].dt.month
new_price['day'] = new_price['loadedDate'].dt.day
```

```python
year_list = [2020., 2021.]

#year_list = list(new_price['year'].unique())
new_price = new_price[new_price.year.isin(year_list)]
```

```python
new_price['year'].unique()
```

```python
new_price= new_price[['tripReadId', 'partnerName', 'cargoTonnage', 'asset_type', 'loadedDate', 'source_region','destination_region', 'source', # 'size', 'unit','delivered', 'goodCategory', destination_region','source_region'
                        'destination', 'goodType', 'asset_size', 'distance', 'totalDistance', 'year', 'month', 'currency', 'amount']] #'fleetAsset_unit', 'asset class', 'size.1', 'unit.1', 'type'      
```

```python
new_price.sample(10)
```

```python
replace_values = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December' }
#trips2021 = trips2021.replace({"months": replace_values})
new_price['month'] = new_price['month'].map(replace_values)
```

```python
new_price.asset_size.unique()
```

```python
new_price.min()  #drop distance = 0, cargoTonnage = 0, amount < 100000
new_price = new_price.loc[~((new_price['asset_size'] == 33000) | (new_price['cargoTonnage'] == 33000) | \
                            (new_price['cargoTonnage'] == 0.0) | (new_price['amount'] <= 1000) | \
                           (new_price['distance'] <= 1)),:]
```

```python
new_price['asset_type'] = new_price['asset_type'].replace({'covered':'Covered', '2x flatbed':'Flatbed'})
new_price['goodType'] = new_price['goodType'].replace({'OIL/LUBE':'OIL AND GAS', 'CASHEW':'CASHEW NUTS', 'CHEMICAL RAW MATERIAL':'CHEMICAL RAW MATERIALS',\
                                         'CHEMICAL':'CHEMICAL RAW MATERIAL', 'OTHER': 'OTHERS', 'SEMO': 'CORN FLOUR', 'FLOUR':'FLOUR',\
                                        'BEVERAGES':'BEVERAGES AND DRINKS', 'EQUIPMENT': 'EQUIPMENTS', 'FERTILIZER': 'AGROCHEMICAL PRODUCTS',\
                                        'HOUSEHOLD AND OFFICE' : 'HOUSEHOLD ITEMS (AC, FURNITURES,UTENSILS, ETC)', 'AGRO': 'AGROCHEMICAL PRODUCTS'})
```
